```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
raw_mail= pd.read_csv('/content/mail_data (1).csv')
```

```python
raw_mail.head()
```

```python
raw_mail.info()
```

```python
mail_data=raw_mail.where(pd.notnull(raw_mail),'')
```

```python
mail_data.shape
```

```python
mail_data['Category'].replace({'ham':0, 'spam':1},inplace=True )
```

```python
mail_data.head()
```

```python
X= mail_data['Message']
y= mail_data['Category']
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.2, random_state=3)
```

```python
feature_extraction= TfidfVectorizer(min_df=1,stop_words='english', lowercase=True)
```

```python
X_train_features= feature_extraction.fit_transform(X_train)
```

```python
X_test_features= feature_extraction.transform(X_test)
```

```python
y_train= y_train.astype(int)
y_test=y_test.astype(int)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train_features,y_train)
```

```python
predictions_trained= model.predict(X_train_features)
```

```python
acc_training_data= accuracy_score(y_train, predictions_trained)
acc_training_data
```

```python
prediction_test=model.predict(X_test_features)
```

```python
acc_test_data= accuracy_score(y_test, prediction_test)
acc_test_data
```
